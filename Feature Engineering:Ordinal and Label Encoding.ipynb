# Encoding: Ordinal and Label

```python
#Import Reqired liberaries
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/kaggle/input/udacity-course-catalog/all_courses.csv",usecols=['Title','Level','Type','Rating'],nrows=100)
df.sample(10)
```

```python
df.iloc[:,1:]
```

# 1. Train-Test-Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(df.iloc[:,[2]], df.iloc[:,[1]],test_size=0.3)
```

```python
X_train
```

# 2. Ordinal Encoding

```python
# For Ordinal Categorical Data - having order
```

```python
from sklearn.preprocessing import OrdinalEncoder  #Step 1: Import
```

```python
oe=OrdinalEncoder(categories=[['beginner','intermediate','advanced']])   #Step 2: Create object
```

```python
oe.fit(X_train)  #Step 3: Fit train data in object
```

```python
X_train=oe.transform(X_train)    #step 4: Transform Train
X_test=oe.transform(X_test)      #And Test data
```

```python
X_train
```

# 3. Label Encoding

```python
# For Label/Output Column
```

```python
from sklearn.preprocessing import LabelEncoder  #Step 1: Import
```

```python
le=LabelEncoder()    #Step 2: Create object
```

```python
le.fit(y_train)    #Step 3: Fit train data in object
```

```python
#check classes
le.classes_
```

```python
#Step 4: Transformion
y_train=le.transform(y_train)   # train data
y_test=le.transform(y_test)    # test data
```
